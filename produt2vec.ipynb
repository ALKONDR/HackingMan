```python
shops = np.array([])

for uid in uids:
    indexes = np.where(df["good_id"] == uid)[0]
#     print(indexes)
    uid_shops = []
    for index in indexes:
        uid_shops.append(df["shop_id"][index])
#         print(index)
#         print(df["item_name"][index])
        
    shops = np.append(shops, np.unique(uid_shops))

shops = np.unique(shops)
shops = shops.astype(int)
print(shops)
```
